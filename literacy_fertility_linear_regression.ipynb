As a confirmation of the linear relationship between the two variables, the Pearson coefficient was calculated. A value of 0.80 was determined, which is quite high.

```python
# Function for Pearson coefficient given two arrays
def pearson_r(x, y):
    # Compute correlation matrix and return entry [0,1]
    corr_mat = np.corrcoef(x, y)
    return corr_mat[0,1]

# Compute Pearson correlation coefficient for illiteracy vs. fertility
print(pearson_r(illiteracy, fertility))
```

Linear regression was then performed to determine the slope and intercept for fertility vs. illiteracy.

```python
# Perform linear regression
a, b = np.polyfit(illiteracy, fertility, 1)
print('slope =', a, 'children per woman / percent illiterate')
print('intercept =', b, 'children per woman')

# Compute regression line
x = np.array([0, 100])
y = a*x + b

# Plot the illiteracy rate versus fertility with regression line
plt.plot(illiteracy, fertility, marker='.', linestyle='none')
plt.margins(0.02)
plt.plot(x, y) # Add regression line to your plot
plt.xlabel('Illiteracy (%)')
plt.ylabel('Fertility')
plt.show()
```

Demonstration that the optimal value of slope was found by np.polyfit() is shown below.

```python
# Specify slopes to consider in the range of 0 to 0.1 (optimal found above is around 0.05)
a_vals = np.linspace(0, 0.1, 200)

# Initialize array for sum of square of residuals
rss = np.empty_like(a_vals)

# Compute sum of square of residuals for each value of a_vals
for i, a in enumerate(a_vals):
    rss[i] = np.sum((fertility - a*illiteracy - b)**2)

# Plot the RSS
plt.plot(a_vals, rss, '-')
plt.xlabel('Slope (children per woman / % illiterate)')
plt.ylabel('RSS')
plt.show()
```
